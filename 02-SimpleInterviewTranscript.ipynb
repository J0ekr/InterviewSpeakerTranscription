```python
print("Loading the model")
model = whisper.load_model("large-v2")
```

```python
def transcribeAudio(audio_file):
    transcription = model.transcribe(audio_file)
    return transcription
```

```python
# A function to timestamp the speech segments:
def convert(seconds):
    h = int(seconds // 3600)
    m = int((seconds % 3600) // 60)
    s = int(seconds % 60)
    return f"{h:02d}:{m:02d}:{s:02d}"
 
# Path to save the interview transcript in .txt format

 
# Save the transcript
def saveTranscript(interview_transcript_path, transcription):
    with open(interview_transcript_path , 'w', encoding='utf-8') as f:
        for segment in transcription["segments"]:
            start_time = convert(segment['start'])
            end_time = convert(segment['end'])
            f.write(f"{start_time} - {end_time}: {segment['text']}")
```

```python
for fileending in types:
    pathlist = Path(data).glob(fileending)
    # print(fileending, len(list(pathlist)))
    
    for audio_file in pathlist:
        f_name = f"{str(audio_file).split('.')[0].split('/')[1]}.txt"
        print(f_name)

        transcript = transcribeAudio(str(audio_file))
        interview_transcript_path = f"{dataOut}/{f_name}"
        saveTranscript(interview_transcript_path, transcript)
```
